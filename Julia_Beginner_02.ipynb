# Install Plot package
Julia does not fix the drawing package like Python's matplotlib.
It seems that we can choose various package combinations at the present stage.

This time, I selected "Plots" as the drawing package and installed it as follows.
The OS is macOS 10.14 (Mojave).

First, start REPL.

Enter "]" while the prompt is shown as "julia>".

Then the prompt will be shown as "(v1.1) pkg>", enter "add Plots".

---
Example of terminal screen

julia> # input]

(v1.1) pkg> add Plots

---

The process may take some time, but when it is finished, ends the REPL.

# Load package
After completing the installation of the "Plots" package, load the package.


---
Example of terminal screen

julia> using Plots

julia> gr () # Plots.GRBackend ()

---

Now you can plot with "Plots"in front end and GR in back end.

### Ex.1 Plot random number data -Line-

```julia
using Plots # Load frontend package
gr()        # Load backend package

plot(randn(30,1))
```

```julia
savefig("plot1.png")    # Save figure as "plot1.png"
```

### Ex.4 Plot Sigmoid function

```julia
x = -5.0:0.1:5.0 # Range of x
f(x) = 1 ./ (1 .+ exp.(-x)) # Sigmoid function

y = f(x)

plot(x, y)
```

```julia
savefig("Sigmoid.png")   # Save figure as "Sigmoid.png"
```
